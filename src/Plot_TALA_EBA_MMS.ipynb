```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
incomp_path_tol_1e_6 = "/Users/jgra0019/Documents/codes/uw3-dev/local-code/out_files/incomp_ksp_tol_1e-6.csv"
incomp_path_tol_1e_8 = "/Users/jgra0019/Documents/codes/uw3-dev/local-code/out_files/incomp_ksp_tol_1e-8.csv"
incomp_path_tol_1e_10 = "/Users/jgra0019/Documents/codes/uw3-dev/local-code/out_files/incomp_ksp_tol_1e-10.csv"


incomp_df_1e_6 = pd.read_csv(incomp_path_tol_1e_6)
incomp_df_1e_8 = pd.read_csv(incomp_path_tol_1e_8)
incomp_df_1e_10 = pd.read_csv(incomp_path_tol_1e_10)

tala_path_tol_1e_6 = "/Users/jgra0019/Documents/codes/uw3-dev/local-code/out_files/tala_ksp_tol_1e-6.csv"
tala_path_tol_1e_8 = "/Users/jgra0019/Documents/codes/uw3-dev/local-code/out_files/tala_ksp_tol_1e-8.csv"
tala_path_tol_1e_10 = "/Users/jgra0019/Documents/codes/uw3-dev/local-code/out_files/tala_ksp_tol_1e-10.csv"
tala_df_1e_6 = pd.read_csv(tala_path_tol_1e_6)
tala_df_1e_8 = pd.read_csv(tala_path_tol_1e_8)
tala_df_1e_10 = pd.read_csv(tala_path_tol_1e_10)
```

```python
# dPdx and dPdz
fig, ax = plt.subplots(2, 2, dpi = 300)

ax[0, 0].set_title(r"$v_x$")
ax[0, 0].plot(1/incomp_df_1e_6["RES"], incomp_df_1e_6["Vx_RMSE"], marker = "o", markersize = 2, 
              linestyle = "-", color = "k", linewidth = 1,
              label = "Incompressible")
# ax[0, 0].plot(1/incomp_df_1e_8["RES"], incomp_df_1e_8["Vx_RMSE"], marker = "o", markersize = 2, linestyle = "-", color = "C0", linewidth = 1,label = "tol = 1e-8 (incompressible)")
# ax[0, 0].plot(1/incomp_df_1e_10["RES"], incomp_df_1e_10["Vx_RMSE"], marker = "o", markersize = 2, linestyle = "-", color = "C1", linewidth = 1,label = "tol = 1e-10 (compressible)")
ax[0, 0].plot(1/tala_df_1e_6["RES"], tala_df_1e_6["Vx_RMSE"], marker = "o", markersize = 2, 
              linestyle = "--", color = "k", linewidth = 1,
              label = "Compressible")
# ax[0, 0].plot(1/tala_df_1e_8["RES"], tala_df_1e_8["Vx_RMSE"], marker = "o", markersize = 2, linestyle = "--", color = "C0", linewidth = 1,label = "tol = 1e-8 (TALA)")
# ax[0, 0].plot(1/tala_df_1e_10["RES"], tala_df_1e_10["Vx_RMSE"], marker = "o", markersize = 2, linestyle = "--", color = "C1", linewidth = 1,label = "tol = 1e-10 (TALA)")
ax[0, 0].set_yscale("log")
ax[0, 0].set_xscale("log")
ax[0, 0].legend(fontsize = 8, loc = 2)

ymin = 0.5*min(tala_df_1e_6["Vx_RMSE"].min(), tala_df_1e_6["Vz_RMSE"].min(), 
           incomp_df_1e_6["Vx_RMSE"].min(), incomp_df_1e_6["Vz_RMSE"].min())
ymax = 2.*max(tala_df_1e_6["Vx_RMSE"].max(), tala_df_1e_6["Vz_RMSE"].max(), 
                    incomp_df_1e_6["Vx_RMSE"].max(), incomp_df_1e_6["Vz_RMSE"].max())
ax[0, 0].set_ylim(ymin, ymax)

ax[0, 1].set_title(r"$v_z$")
ax[0, 1].plot(1/incomp_df_1e_6["RES"], incomp_df_1e_6["Vz_RMSE"], marker = "o", markersize = 2, 
              linestyle = "-", color = "k", linewidth = 1,
              label = "Incompressible")
ax[0, 1].plot(1/tala_df_1e_6["RES"], tala_df_1e_6["Vz_RMSE"], marker = "o", markersize = 2, 
              linestyle = "--", color = "k", linewidth = 1,
              label = "Compressible")
ax[0, 1].set_yscale("log")
ax[0, 1].set_xscale("log")
ax[0, 1].set_ylim(ymin, ymax)


ax[1, 0].set_title(r"$\frac{\partial P}{\partial x}$")
ax[1, 0].plot(1/incomp_df_1e_6["RES"], incomp_df_1e_6["dPdx_RMSE"], marker = "o", markersize = 2, 
              linestyle = "-", color = "k", linewidth = 1,
              label = "Incompressible")
ax[1, 0].plot(1/tala_df_1e_6["RES"], tala_df_1e_6["dPdx_RMSE"], marker = "o", markersize = 2, 
              linestyle = "--", color = "k", linewidth = 1,
              label = "Compressible")
ax[1, 0].set_yscale("log")
ax[1, 0].set_xscale("log")

ymin = 0.5*min(tala_df_1e_6["dPdx_RMSE"].min(), tala_df_1e_6["dPdz_RMSE"].min(), 
                    incomp_df_1e_6["dPdx_RMSE"].min(), incomp_df_1e_6["dPdz_RMSE"].min())
ymax = 2.*max(tala_df_1e_6["dPdx_RMSE"].max(), tala_df_1e_6["dPdz_RMSE"].max(), 
                    incomp_df_1e_6["dPdx_RMSE"].max(), incomp_df_1e_6["dPdz_RMSE"].max())

ax[1, 0].set_ylim(ymin, ymax)

ax[1, 1].set_title(r"$\frac{\partial P}{\partial z}$")
ax[1, 1].plot(1/incomp_df_1e_6["RES"], incomp_df_1e_6["dPdz_RMSE"], marker = "o", markersize = 2, 
              linestyle = "-", color = "k", linewidth = 1,
              label = "Incompressible")
ax[1, 1].plot(1/tala_df_1e_6["RES"], tala_df_1e_6["dPdz_RMSE"], marker = "o", markersize = 2, 
              linestyle = "--", color = "k", linewidth = 1,
              label = "Compressible")
ax[1, 1].set_yscale("log")
ax[1, 1].set_xscale("log")
ax[1, 1].set_ylim(ymin, ymax)

for a in ax.flat:
    a.set_xlabel("Element size")
    a.set_ylabel(r"$L_2$ norm")

# Hide x labels and tick labels for top plots and y ticks for right plots.
for a in ax.flat:
    a.label_outer()

fig.tight_layout()

# ax.plot(1/incomp_df["RES"], incomp_df["dPdx_RMSE"],  marker = "o", linestyle = "dashdot", color = "k", linewidth = 1, markersize = 2,label = r"$\frac{\partial p}{\partial x}$ incompressible")
# ax.plot(1/incomp_df["RES"], incomp_df["dPdz_RMSE"],  marker = "o", linestyle = "dotted", color = "k", linewidth = 1, markersize =2, label = r"$\frac{\partial p}{\partial z}$ incompressible")
# ax.plot(1/comp_df["RES"], comp_df["dPdx_RMSE"],  marker = "o", linestyle = "dashdot", color = "C0", linewidth = 1, markersize = 2,label = r"$\frac{\partial p}{\partial x}$ compressible")
# ax.plot(1/comp_df["RES"], comp_df["dPdz_RMSE"],  marker = "o", linestyle = "dotted", color = "C0", linewidth = 1, markersize =2, label = r"$\frac{\partial p}{\partial z}$ compressible")
# ax.set_xlabel("Element size")
# ax.set_ylabel(r"$L_2$ norm")
# ax.set_yscale("log")
# ax.set_xscale("log")
# ax.legend()
# #ax.set_aspect("equal")
# #ax.set_title("Incompressible Stokes")
# ax.set_title("Tolerance = 1e-10")
```
